```
import requests
website_url = requests.get('https://www.booking.com/city/ca/vancouver.en-gb.html?aid=356980;label=gog235jc-1DCAMoJzjZA0gzWANobIgBAZgBCbgBGcgBDNgBA-gBAYgCAagCA7gC4uOO7wXAAgE;sid=b18db7827357945437fc64132cc3217b;inac=0&keep_landing=1&').text
```

```
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```
My_table = soup.find('div',{'class':'sr__card_content'})
My_table
```

```
links = My_table.findAll('a')
links
```

```
Countries = []
for link in links:
    Countries.append(link.get('href'))
    
print(Countries)
```

```
import pandas as pd
df = pd.DataFrame()
df['Country'] = Countries

df
```
